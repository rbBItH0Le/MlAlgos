```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Mall_Customers.csv')
```

```python
df
```

```python
df.isna().sum()
```

```python
sex=pd.get_dummies(df['Genre'],drop_first=True)
```

```python
sex
```

```python
X=pd.concat([df,sex],axis=1)
```

```python
X
```

```python
X.drop(['CustomerID','Genre'],axis=1,inplace=True)
```

```python
X
```

```python
from sklearn.cluster import DBSCAN
```

```python
db=DBSCAN(eps=12,min_samples=4)
```

```python
db.fit(X)
```

```python
du=pd.DataFrame(db.labels_)
```

```python
X=pd.concat([X,du],axis=1)
```

```python
X
```

```python
from sklearn.metrics import silhouette_score
```

```python
X.drop([0],axis=1,inplace=True)
```

```python
X
```

```python
score=silhouette_score(X,db.labels_)
```

```python
score
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
```

```python
df.head()
```

```python
Sex=pd.get_dummies(df['Sex'],drop_first=True)
```

```python
Sex
```

```python
Embarked=pd.get_dummies(df['Embarked'],drop_first=True)
```

```python
Embarked
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
g=pd.concat([df,Sex,Embarked],axis=1)
```

```python
g.head()
```

```python
g.tail()
```

```python
f=g.dropna()
```

```python
f.head()
```

```python
f.drop(['Cabin'],axis=1,inplace=True)
```

```python
f.head()
```

```python
f.drop(['Sex','Embarked'],axis=1,inplace=True)
```

```python
f.head()
```

```python
y=f['Survived']
```

```python
y
```

```python
X=f.iloc[:,1:]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
ff=rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
ac=accuracy_score(ff,y_test)
```

```python
ac
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters=[{'max_features':['sqrt','log2','auto'],'criterion':['gini','entropy'],'n_estimators':[10,20,30,40,60,80,100],'min_samples_split':[2,3,4,5]}]
```

```python
grid_search=GridSearchCV(estimator=rf,param_grid=parameters,cv=10,n_jobs=-1)
```

```python
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```
